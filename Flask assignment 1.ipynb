### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask , if you want to create a link for a specific page we use the url_for() function . This function helps us to create a correct web address for that page.even if you change things around later. It ensure that links always work correctly

```python
from Flask import Flask url_for
app = Flask(_name_)

@app.route("/")
def home():
    return "Welcome to the home page"

@app.route("/about")
def about():
    return "This is the about page"


@app.route("/contact")
def contact():
    return "You can reach us at contact@example.com"

if __name__ =='__main__':
    with app.test_request_context():
        print("URL for home:" , url_for('home'))
        print("URL for about:" , url_for('about'))
        print("URL for contact:" , url_for('contact'))
```
