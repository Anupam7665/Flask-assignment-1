### Q1. Explain GET and POST methods.

The GET method is primarily used to request data from a  specified resources . 
Data is appended to the URL as query parameters. GET request can be cached amd its store in the browser history . GET data is passed through URL , this making it less secure for sensitive information  for eg _Password_


The POST is used to send data to a specific palce . It is like filling a form and hitting a submit button on it . This data can be creating , updating , deleting on the server . It has no limit how much data you have to send. Data is sent in the requseet body which is not visible in the URL. POST request are generally not cached. POST is cosnidered is more secure for sensitive data as it doesn't expose data in the URL

### Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to send a response to a web browser.
It allows you to move user from one page of your website to another . After someone submits a form on your website you can use _redirect()_ to send them to _"thankyou page"_ instead to them to showing the same form again.
It's useful for creating clean and user-friendly URLs . When there's an error or a page doesn't exist you can use _redirect_() to guide user to an error page or error message
